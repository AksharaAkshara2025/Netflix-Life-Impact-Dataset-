# Minute of Life-Changing Insight

```python
netflix['minute_of_life-changing_insight']
```

```python
import re   #regular Expression

def splitInsights(insight):
    insight = str(insight).strip() # Converting the text(insights) into string
    insight = re.sub(r'[–—_]', '-', insight) #It is maintaing the uniformity of matching the patterns by removing all double dashes with -
    match = re.match(r'(?:(\d+):)?(\d+):(\d+)\s*-\s*(.*)', insight)
    if match:
        hours = int(match.group(1)) if match.group(1) else 0  # Default to 0 if no hours provided
        minutes = int(match.group(2))
        seconds = int(match.group(3))
        description = match.group(4).strip()

        totalSeconds = hours * 3600 + minutes * 60 + seconds
        return pd.Series([totalSeconds, description])
    return pd.Series([np.nan, None])


```

```python
netflix[['insight_sec', 'insight_dec']] = netflix['minute_of_life-changing_insight'].apply(splitInsights)
```

```python
netflix[['insight_sec', 'insight_dec']]
```

H0: There is no significant effect on the average_rating of the movie release before 2015

H1: There is a significant relationship on the average_rating and movie released after 2015

```python
from pickle import FALSE
from scipy import stats

netflix['post2015'] = netflix['release_year'] > 2015

before_2015 = netflix[netflix['post2015'] == False]['average_rating']
after_2015 = netflix[netflix['post2015'] == True]['average_rating']


netflix['post2015']
```

### T-Tests

1. One sample: meand of sample with population

2. Two Sample dependent: mean of two samples (paired sample>> before and after)

3. Two independent samples: mean of two samples

```python
t_stat, p_value = stats.ttest_ind(before_2015, after_2015)
print(f't_stat: {t_stat}, p-value: {p_value}')
```

### Cohen D tests


```python
import numpy as np

mean_diff = before_2015.mean() - after_2015.mean()
pooled_std = np.sqrt((before_2015.std()**2 + after_2015.std()**2) / 2)

cohens_d = mean_diff/pooled_std

print(f"Cohen's D: {cohens_d}")
```
